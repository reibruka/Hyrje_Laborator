## 1. Të dhënat bruto

Vendosni të dhënat tuaja këtu.  
**Njësia** duhet të jetë e qartë (mm, s, g, etj.).

> Shembulli më poshtë përmban matje të reja për këtë sesion.


```python
import numpy as np
import matplotlib.pyplot as plt

# Të dhënat e reja për Rei Bruka
data = np.array([
    12.5, 12.7, 12.4, 12.6, 12.8,
    12.5, 12.6, 12.4, 12.3, 12.5,
    12.7, 12.6, 12.5, 12.4, 12.6
])

unit = "cm"  
data
```

## 2. Statistikat bazë

Përdorni:
- `np.mean(data)` për mesataren
- `np.std(data, ddof=1)` për devijimin standard të kampionit

Formulat:

$$
\bar{x} = \frac{1}{N}\sum_{i=1}^{N}x_i
$$

$$
s = \sqrt{\frac{1}{N-1}\sum_{i=1}^{N}(x_i-\bar{x})^2}
$$

Gabimi i mesatares:

$$
\sigma_{\bar{x}} = \frac{s}{\sqrt{N}}
$$


```python
N = len(data)
xbar = np.mean(data)
s = np.std(data, ddof=1)
sem = s / np.sqrt(N)

N, xbar, s, sem
```

```python
# Raportimi i rezultatit
print(f"Rezultati: {xbar:.3f} ± {sem:.3f} {unit}")
```

## 3. Histogrami

Histogrami tregon shpërndarjen e matjeve.

Kërkesa minimale për grafik:
- etiketa boshtesh + njësi
- titull
- grid

```python
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.xlabel(f"Vlera ({unit})")
plt.ylabel("Frekuenca")
plt.title("Histogrami i matjeve - Rei Bruka")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 4. Grafiku i matjeve sipas rendit

Ky grafik ndihmon të identifikoni:
- prirje (drift) gjatë matjeve
- matje të dyshimta (outliers)

```python
plt.plot(data, 'o-', color='darkorange')
plt.xlabel("Numri i matjes")
plt.ylabel(f"Vlera ({unit})")
plt.title("Ecuria e matjeve")
plt.grid(True)
plt.show()
```
